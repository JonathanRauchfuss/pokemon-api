```python
# We only want the Driver Standings
# were basically going down a big list item by item until we get to our driver standings 
# the 0 represents when we get to the "season" list we have to specify because its a list where in the list do we want to look 
# we can also create a for loop to ask for a certain amount of items within a dictionary 
# for x in range(5):
    # data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"][x]["Driver"]["familyName"]
    # print(data) 

my_racer_data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
my_racer_data
```

```python
def get_racer_info(data): # were crreating a function 
    new_data=[]    #were creating a new list and each dictionary we create will go inside this list 
    for racer in data:    # for each dictionary that gets passed in, in a list full of dictionaries 
        racer_dict = {}    # were making a new dictionary 
        racer_name = f'{racer["Driver"]["givenName"]} {racer["Driver"]["familyName"]}' #were getting the first and last name from the first dictionary in the list of dictionaries that were passing in 
        racer_dict[racer_name]={  # were filling in the ductionary we created #technically the racer name is the key and all of the other information are values but becuase all of the values are in another dictionary we can create key value pairs 
            "first_name":racer["Driver"]["givenName"], # we are looking through the first dictionary we passed in and are grabbing the data, if something is nested within a dictionary or list we have to choose the parent and then the child, so for this example the given name is nested inside the Driver dictionary so we first look for the driver dictionary then we look for the given name 
            "last_name":racer["Driver"]["familyName"],
            "position":racer["position"],
            "wins":racer["wins"],
            "DOB":racer["Driver"]["dateOfBirth"],
            "Nationality":racer["Driver"]["nationality"],
            "constructor":racer["Constructors"][0]["name"] # here we have a list after constructors so after we go in the constructors list we specify which position we want to start in the list then we specify the item we want in the list 
        }
        new_data.append(racer_dict)  # here we are adding the dictionary we made to the new data list 
    return new_data

get_racer_info(my_racer_data)
```

check out this Pokemon API https://pokeapi.co/
Use the requests package to connect to this API and get and store data for 20 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny', attack base_state, hp base_stat, defense base_stat

```python
def get_pokemon_data(pokemon_id):
    url = f'https://pokeapi.co/api/v2/pokemon/{pokemon_id}'
    response = requests.get(url)

    if response:
        data = response.json()
        return {
            'name': data['name'],
            'ability': data['abilities'][0]['ability']['name'],
            'base_experience': data['base_experience'],
            'sprite_url': data['sprites']['front_default'],
        }
    else:
        print(f"Cant get data for Pokemon ID {pokemon_id}")
        


pokemon_data_list = []
for i in range(1, 21):  
    pokemon_data = get_pokemon_data(i)
    if pokemon_data:
        pokemon_data_list.append(pokemon_data)


for pokemon_data in pokemon_data_list:
    print(f"Name: {pokemon_data['name']}")
    print(f"Ability: {pokemon_data['ability']}")
    print(f"Base Experience: {pokemon_data['base_experience']}")
    print(f"Sprite URL: {pokemon_data['sprite_url']}")
    print("\n")
```
